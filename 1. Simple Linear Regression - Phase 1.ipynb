```python
import pandas as pd # Import the pandas library that helps you work with data in tables and give it the alias 'pd'
dataset=pd.read_csv("Salary_Data.csv") # Read the CSV file named "Salary_Data.csv" into a dataframe called 'dataset'
```

```python
dataset # Print the rows of the dataset 
```

```python
independent=dataset[["YearsExperience"]] # Assign the input variable  'YearsExperience' column to the variable named  'independent'
```

```python
independent # Print the rows of the independent 
```

```python
dependent=dataset[["Salary"]] # Assign the output  'Salary' column to the variable named  'dependent'
```

```python
dependent # Print the rows of the dependent 
```

```python
from sklearn.model_selection import train_test_split #Import the train_test_split function from the sklearn.model_selection module
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) #Split data into 80% train and 20% test
```

```python
from sklearn.linear_model import LinearRegression # import the LinearRegression class from sklearn module
regressor = LinearRegression() # Create an instance of the LinearRegression model
regressor.fit(x_train,y_train) # Train the model using the training data
```

```python
weight=regressor.coef_ # Get the coefficients
weight # Display the weights
```

```python
bias = regressor.intercept_ #Get the intercept
bias # Show the bias
```

```python
y_pred=regressor.predict(x_test) # Predict the target values using the test data
```

```python
from sklearn.metrics import r2_score # Import the r2_score function from sklearn.metrics
r_score = r2_score(y_test, y_pred)  # Calculate the R-squared value
r_score  # Show the R-squared score
```

```python
import pickle # Import the pickle module for saving and loading models
filename ="final_model_linear.sav"  # Name of the file to save the model
pickle.dump(regressor, open(filename,'wb')) # Save the model to the file
loaded_regressor = pickle.load(open("final_model_linear.sav","rb"))  # Load the model from the file
result = loaded_regressor.predict([[1]])  # Make a prediction for input value 13
result  # Show the prediction result

```
