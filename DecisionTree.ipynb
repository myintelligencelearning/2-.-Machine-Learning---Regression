```python
 import pandas as pd # Import the pandas library that helps you work with data in tables and give it the alias 'pd'
```

```python
dataset=pd.read_csv("50_Startups.csv") # Read the CSV file named "50_Startups.csv" into a dataframe called 'df'
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(int) #Convert categorical 'State' column to one-hot encoding column and  drop the first category
```

```python
dataset # Show the output
```

```python
dataset.columns
```

```python
ins=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']] # Assign the input variable columns to the variable named  'inp'
```

```python
ops=dataset[['Profit']] # Assign the output  'Profit' column to the variable named  'op'
```

```python
ops
```

```python
from sklearn.model_selection import train_test_split #Import the train_test_split function from the sklearn.model_selection module
x_train,x_test,y_train,y_test=train_test_split(ins,ops,test_size=0.3,random_state=0) #Split data into 70% train and 30% test
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor #Import the SVR (Support Vector Regression) model class from scikit-learn
regressor=DecisionTreeRegressor() # Create an SVR model using a polynomial kernel and regularization parameter C=1000
regressor=regressor.fit(x_train,y_train) #Train the SVR model using the training data
```

```python
import matplotlib.pyplot as plt  # Imports the matplotlib library for plotting graphs
from sklearn import tree  # Imports the tree module from scikit-learn 
tree.plot_tree(regressor)  # Plots the structure of the decision tree 
plt.show()  # Displays the plotted tree 
```

```python
y_pred=regressor.predict(x_test) # Predict the target values using the test data
```

```python
from sklearn.metrics import r2_score  # Import the r2_score function from sklearn.metrics
r_score=r2_score(y_test,y_pred) # Calculate the R-squared value
```
